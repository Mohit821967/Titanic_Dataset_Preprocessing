```python
# prompt: suggest a plot for age column

import matplotlib.pyplot as plt
import seaborn as sns

# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(data['Age'], bins=20, kde=True)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Age')
plt.show()

# Box plot

plt.figure(figsize=(10, 6))  # Adjust the figure size
sns.boxplot(x='Survived', y='Age', data=data,
            palette='Set3',  # Change the color palette
            width=0.5,  # Adjust the width of the boxes
            notch=True)  # Add notches to show the confidence interval around the median
plt.xlabel('Survived', fontsize=12)  # Change font size of x-axis label
plt.ylabel('Age', fontsize=12)  # Change font size of y-axis label
plt.title('Age Distribution by Survival', fontsize=14)  # Change font size of title
plt.grid(True, linestyle='--', alpha=0.7)  # Add grid with custom style
plt.xticks([0, 1], ['Did Not Survive', 'Survived'], fontsize=11)  # Custom x-axis tick labels
plt.yticks(fontsize=11)  # Change font size of y-axis ticks
plt.show()
```
